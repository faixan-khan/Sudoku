```python
def clean_row(possible_value, solution, row_num):

    #Isolate this row
    exist = set(solution[row_num - 1])

    #Clean
    for j in range(1, SIZE_N * SIZE_N + 1):
        # Update the markup for each cell in this row by deleting values that are already in this row
        if (row_num, j) in possible_value:
          possible_value[row_num, j] = [x for x in possible_value[row_num, j] if x not in exist]
```

```python
def clean_col(possible_value, solution, col_num):

    # Isolate this column
    exist = set([x[col_num - 1] for x in solution])
    for i in range(1, SIZE_N * SIZE_N + 1):
        # Update markup by deleting existing numbers from the cells 
        if (i, col_num) in possible_value:
          possible_value[i, col_num] = [x for x in possible_value[i, col_num] if x not in exist]
```

```python
def clean_box(possible_value, solution, box_num):

    #Isolate this box
    box_i = box_range(box_num[0])
    box_j = box_range(box_num[1])

    # Get a list of filled cells in a box
    exist = []
    for i in box_i:
        for j in box_j:
            exist += [solution[i - 1][j - 1]]

    #Clean the box
    for k in box_i:
        for l in box_j:
            if (k, l) in possible_value:
                possible_value[k, l] = [b for b in possible_value[k, l] if b not in exist]
    
    
```

```python
def box_range(number):
    """
    :param number: input the row or column number
    :return: a list of row or column number within the same box
    """
    start = (number - 1)//SIZE_N * SIZE_N + 1
    return list(range(start, start + SIZE_N))
```

```python
def check_unique_possible_value(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    For each cell, if there is only one possible number, update solution_ and remove from possible_value_
    """
    mark_up_copy = copy.deepcopy(possible_value_)

    #Dealing with hidden single
    for key, value in mark_up_copy.items():
        if len(value) == 1: #If a cell can have only one value
            #Enforce this value
            solution_[key[0] - 1][key[1] - 1] = value[0]
            #possible_value_[key] = []
            possible_value_.pop(key)

            #Update row, col and box
            clean_row(possible_value_, solution_, key[0])
            clean_col(possible_value_, solution_, key[1])
            clean_box(possible_value_, solution_, key)

    return 0
```

```python
def loop_basic_rule(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    Run all basic rules until there is no update on solution_
    """
    while True:
        solution_old = copy.deepcopy(solution_)
        check_row(possible_value_, solution_)
        check_column(possible_value_, solution_)
        check_box(possible_value_, solution_)
        check_unique_possible_value(possible_value_, solution_)
        if solution_ == solution_old:
            break
```

```python
#Improvement on the algorithm to find p-sets better
def algorithm(possible_list_, possible_value_, solution_):
    """
    :param possible_list_: list of markups for cells corresponding to row, col, or box
    :param possible_value_: mark up of the whole puzzle
    :param solution_: the list of existing solution
    The function applies Crook's algorithm and eliminate impossible numbers.
    If there is any update, re-run all basic rules
    """
    
    try:
        # Minimum and maximum mark up size for the row, col or box
        min_num_possible = min((len(v)) for _, v in possible_list_.items())
        max_num_possible = max((len(v)) for _, v in possible_list_.items())

        if max_num_possible > SIZE_N * SIZE_N - 2:
            max_num_possible = SIZE_N * SIZE_N - 2

    except ValueError:
        return 0
        
    # We start with the smalles possible p-set for a given big p-set
    for i in range(min_num_possible, max_num_possible + 1):
        #Start with the largest possible combination 
        candidates = {key: value for key, value in possible_list_.items() if len(value) <= i}
        p_set_size = i
        
        #Find p-set of corresponding size 
        p_set = find_p_set(candidates, p_set_size)

        #If p_set is not None
        if p_set:
            changed = False
            #Numbers in this p-set
            values = set([val for y in [v for k, v in candidates.items() if k in p_set] for val in y])
            for key_2, value_2 in {key: value for key, value in possible_list_.items() if
                                    key not in p_set}.items():
                #Delete the corresponding values from the other cells
                old_cell_values = value_2
                possible_value_[key_2] = [x for x in value_2 if x not in values]

                #Observe this run changed anything in the markup
                if len(old_cell_values) != len(possible_value_[key_2]):
                    changed = True

            #If markup changed run the basic check loop
            if changed:
                loop_basic_rule(possible_value_, solution_)
    return 0
```

```python
#Find p-set of the given size among the candidates
def find_p_set(candidates, p_set_size):
  """ 
  :param candidates: dict of cells, corresponding to candidates 
  :param p_set_size: the size of the p-set being searched 
  :return: list of keys that correspond to the found p-set, if such exists else none
  """
  keys = [k for k, _ in candidates.items()]
  
  return find_combs({}, p_set_size, keys, candidates, p_set_size)
```

```python
#Recursive helper function to find p-sets from p-set candidates and number of 
#cells to include in the p-set
def find_combs(prev_cells, current_size, next_cells, candidates, p_set_size):
  """
  :param prev_cells: already formed combination 
  :param current_size: int, needed for halt condition
  :next_cells: the cells to consider 

  """
  #The current set we are working with
  vals = [] 
  for _, v in prev_cells.items():
    vals += v

  #If it is the last element considered (e.g. third for p-set of size 3)
  if current_size == 1:
    for key in next_cells:
      #The p-set size must be satisfied
      if len(set(vals + candidates[key])) == p_set_size:
        toReturn = [k for k, _ in prev_cells.items()]
        toReturn.append(key)
        return toReturn
    #If no p-set of the required size, return none
    return None
  
  else:
    for i, key in enumerate(next_cells):
      #Stop when no options left
      if i == (len(next_cells) - current_size + 1):
        break

      #The p-set size must be satisfied
      if len(set(vals + candidates[key])) > p_set_size:
        continue

      #Study the next cell
      else:
        pass_cells = copy.deepcopy(prev_cells) 
        pass_cells[key] = candidates[key]

        comb = find_combs(pass_cells, current_size - 1, next_cells[i+1: ], candidates, p_set_size)
        if comb:
          return comb
        else:
          continue

    #If the p-set of the given size cannot be satisfed return none
    return None
```

```python
def algorithm_row(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    Apply Crook's algorithm on each row
    """
    for i in range(1, SIZE_N * SIZE_N + 1):
      possible_list = {key: value for key, value in possible_value_.items() if key[0] == i}# and len(value) > 0}

      if len(possible_list) != 0:
        algorithm(possible_list, possible_value_, solution_)

    return 0
```

```python
def algorithm_column(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    Apply Crook's algorithm on each column
    """
    for j in range(1, SIZE_N * SIZE_N + 1):
        possible_list = {key: value for key, value in possible_value_.items() if key[1] == j}# and len(value) > 0}

        if len(possible_list) != 0:
          algorithm(possible_list, possible_value_, solution_)

    return 0
```

```python
def algorithm_box(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    Apply Crook's algorithm on each box
    """
    for i in list(range(1, SIZE_N * SIZE_N, SIZE_N)):
        for j in list(range(1, SIZE_N * SIZE_N, SIZE_N)):
            possible_list = {key: value for key, value in possible_value_.items() if
                             key[0] in box_range(i) and key[1] in box_range(j)}# and len(value) > 0}

            if len(possible_list) != 0:
                algorithm(possible_list, possible_value_, solution_)
    return 0
```

```python
def loop_algorithm(possible_value_, solution_):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    :param solution_: the list of existing solution
    Apply Crook's algorithm until there is no update on solution_
    """
    while True:
        #solution_old = copy.deepcopy(solution_) Need to copy the solution as well? 
        mark_up_old = copy.deepcopy(possible_value_)
        # Row or col in a box - eliminate from others
        check_box_eliminate_others(possible_value_, solution_)
        # Run mutually exclusive
        mutually_exclusive_sets(possible_value_, solution_)
        # Main algorithm - Crook's, find p-sets
        algorithm_row(possible_value_, solution_)
        algorithm_column(possible_value_, solution_)
        algorithm_box(possible_value_, solution_)

        if mark_up_old == possible_value_:
            break

    return 0
```

```python
def check_box_eliminate_others(possible_value_, solution):
    """
    :param possible_value_: the dict of storing all possible numbers of each cell
    By considering the possible numbers within a box, check if there is a possible number only in one row/column.
    If so, then in the same row/column outside the box, this possible number will be eliminated from all markup.
    """
    for i in list(range(1, SIZE_N * SIZE_N, SIZE_N)):
        for j in list(range(1, SIZE_N * SIZE_N, SIZE_N)):
            #Isolate value of this box
            possible_element = [x for b in [value for key, value in possible_value_.items()
                                     if key[0] in box_range(i) and key[1] in box_range(j)] for x in b]

            #Get values with frequency at most 3 otherwise cannot be in the same row or column
            possible_element = [x for x in set(possible_element) if possible_element.count(x) < SIZE_N + 1]

            for x in possible_element:
                #Monitor for changes
                changed = False

                #Get keys of this box that have this value x
                available_cell = [key for key, value in possible_value_.items()
                                  if x in value if key[0] in box_range(i) and key[1] in box_range(j)]
                if len(set([x[0] for x in available_cell])) == 1: #If the cells are on the same row

                    #Then look at the cells of the same row not in this box
                    for key in [key for key, value in possible_value_.items() if
                                key[0] == available_cell[0][0] and key not in available_cell]:
                        #Delete this number from other cells 
                        old_cell = possible_value_[key]
                        possible_value_[key] = [y for y in possible_value_[key] if y != x]

                        #Run the basic check loop if change has been made
                        if len(old_cell) != len(possible_value_[key]):
                            changed = True

                #Do the same with the columns
                if len(set([x[1] for x in available_cell])) == 1:
                    for key in [key for key, value in possible_value_.items() if
                                key[1] == available_cell[0][1] and key not in available_cell]:
                        old_cell = possible_value_[key]
                        possible_value_[key] = [y for y in possible_value_[key] if y != x]

                        #Run the basic check loop if change has been made
                        if len(old_cell) != len(possible_value_[key]):
                            changed = True

                if changed:
                    loop_basic_rule(possible_value_, solution)
    return 0
```

```python
def mutually_exclusive_sets(possible_value_, solution):
  #Finds mutually exclusive sets 
  #You can either find through looking at boxes or rows and columns
  #We do by looking at rows and columns

  #Examine each row
  for row in range(1, SIZE_N * SIZE_N + 1): #change after
      #Get the working dictionary of this row
      row_dict = {k: v for k, v in possible_value_.items() if k[0] == row}
      #list of values to chech the frequencies
      row_values = [v for val in [vv for _, vv in row_dict.items()] for v in val]
      #final_vals will store values with frequency smaller than 4 (2 or 3)
      final_vals = [v for v in set(row_values) if row_values.count(v) < (SIZE_N + 1) and row_values.count(v) > 1]

      changed = False

      #Check if the values are in the same box
      for val in final_vals:
          #Monitor if change is made
          
          #Get the cell keys that have this value
          v_cells = {k: v for k, v in row_dict.items() if val in v} 
          keys = [k for k, _ in v_cells.items()] 
          #Check if all of them are in the same box
          same_box = True
          for i, k in enumerate(keys):
              if i == len(keys) - 1:
                  break
              else:
                  if box_range(k[0]) != box_range(keys[i + 1][0]) or box_range(k[1]) != box_range(keys[i + 1][1]):
                      same_box = False
                      break

          #If they are all in the same box 
          
          if same_box:
              #Look at other cells in that box
              box_cells = {k: v for k, v in possible_value_.items() if k[0] in box_range(keys[0][0]) and k[1] in box_range(keys[0][1])
                           and k not in keys and val in v}

              #Update accordingly, i.e. delete the values that cannot be in those cells
              for k in box_cells:
                  old_cell = possible_value_[k]
                  possible_value_[k] = [v for v in possible_value_[k] if v != val]

                  #Loop basic rule if change has been made 
                  if len(old_cell) != len(possible_value_[k]):
                      changed = True

      if changed:
          loop_basic_rule(possible_value_, solution)
              

  #Do the same with the columns 
  for col in range(1, SIZE_N * SIZE_N + 1): 
      changed = False
      #Get the working dictionary of this col
      col_dict = {k: v for k, v in possible_value_.items() if k[1] == col}
      #list of values to chech the frequencies
      col_values = [v for val in [vv for _, vv in col_dict.items()] for v in val]
      #final_vals will store values with frequency smaller than 4 (2 or 3)
      final_vals = [v for v in set(col_values) if col_values.count(v) < (SIZE_N + 1) and col_values.count(v) > 1]

      #Check if the values are in the same box
      for val in final_vals:
          #Get the cell keys that have this value
          v_cells = {k: v for k, v in col_dict.items() if val in v} 
          keys = [k for k, _ in v_cells.items()] 
          #Check if all of them are in the same box
          same_box = True
          for i, k in enumerate(keys):
              if i == len(keys) - 1:
                  break
              else:
                  if box_range(k[0]) != box_range(keys[i + 1][0]) or box_range(k[1]) != box_range(keys[i + 1][1]):
                      same_box = False
                      break
          #If they are all in the same box 
          if same_box:
              #Look at other cells in that box
              box_cells = {k: v for k, v in possible_value_.items() if k[0] in box_range(keys[0][0]) and k[1] in box_range(keys[0][1])
                           and k not in keys and val in v}

              #Update accordingly, i.e. delete the values that cannot be in those cells
              for k in box_cells:
                  old_cell = possible_value_[k]
                  possible_value_[k] = [v for v in possible_value_[k] if v != val]

                  #Loop basic rule if change has been made 
                  if len(old_cell) != len(possible_value_[k]):
                      changed = True

      if changed:
          loop_basic_rule(possible_value_, solution)
  
  return 0
```

```python
def make_guess(possible_value, solution):
    
    print('\nMade a guess')
    
    #Find a cell with fewest options in the mark up (so ideally 2)
    guess = ()
    smallest = 50 #arbitrary large number
    for k, v in possible_value.items():
        if len(v) == 2:
            guess = k
            break
        elif len(v) < smallest:
            guess = k
            smallest = len(v)

    print('Guessed cell', guess, possible_value[guess])
    print('The current state: ')
    print_sudoku(solution)
    
    #Guess contains the key of the cell guessed on
    #Now paste one value from the mark up of this cell and solve
    for v in possible_value[guess]:
        #Make copies of the solution and the markup, for each guess new copy
        solution_copy = copy.deepcopy(solution)
        mark_up_copy = copy.deepcopy(possible_value)

        #Enforce the guess
        solution_copy[guess[0] - 1][guess[1] - 1] = v
        mark_up_copy.pop(guess)

        #Update the mark up of the copy after enforcement 
        clean_row(mark_up_copy, solution_copy, guess[0])
        clean_col(mark_up_copy, solution_copy, guess[1])
        clean_box(mark_up_copy, solution_copy, guess)

        #Solve with the guessed sell
        result = solver(mark_up_copy, solution_copy, guessed=True)

        #If a valid solution is found return it 
        if result:
            return result

    #If checked all values and no valid solution return none
    return None
```

```python
def consistency_check(possible_value, solution):
    #Based on mark up and current solution check if the solution is guaranteed 
    #to be wrong
    incorrect = False

    #1. Check if there are empty cells in the markup, as there should not be
    #because cells with empty mark up means there is already a number in the cell
    #in which case the key for this cell is deleted
    for k, v in possible_value.items():
        if len(v) == 0:
            incorrect = True
            return incorrect 
    
    #2. Check if there are rows, columns or boxes with fewer mark up values 
    #than the number of cells

    #For rows
    for i in range(1, SIZE_N * SIZE_N + 1):
      possible_list = {key: value for key, value in possible_value.items() if key[0] == i}
      
      if len(possible_list) != 0:
          if len(possible_list) != len(set([v for _, cell in possible_list.items() for v in cell])):
              incorrect = True
              return incorrect
    
    #For columns
    for j in range(1, SIZE_N * SIZE_N + 1):
        possible_list = {key: value for key, value in possible_value.items() if key[1] == j}

        if len(possible_list) != 0:
            if len(possible_list) != len(set([v for _, cell in possible_list.items() for v in cell])):
                incorrect = True
                return incorrect

    #For boxes 
    for i in list(range(1, SIZE_N, SIZE_N * SIZE_N)):
        for j in list(range(1, SIZE_N, SIZE_N * SIZE_N)):
            possible_list = {key: value for key, value in possible_value.items() if
                             key[0] in box_range(i) and key[1] in box_range(j)}

            if len(possible_list) != 0:
                if len(possible_list) != len(set([v for _, cell in possible_list.items() for v in cell])):
                    incorrect = True
                    return incorrect

    return incorrect
```

```python
def solver(possible_value, solution, guessed=False):

    # Step - 1: Run basic check first, enforce if obvious cells exist
    loop_basic_rule(possible_value, solution)
        
    while True:
        # Remember the current state of the puzzle (needed for guessing and stop condition)
        solution_old = copy.deepcopy(solution)
        possible_value_old = copy.deepcopy(possible_value)
      
        # Crooks algorithm 
        loop_algorithm(possible_value, solution)
        
        #If the current solution process is based on a guess
        if guessed:
            #Then check if it is correct so far
            incorrect = consistency_check(possible_value, solution)
            if incorrect:
                # Backtrack if there is inconsistency
                print('The last guess was wrong see solution, backtracking')
                print_sudoku(solution)
                return None

        #If no improvements have been made see if the solution was found or not
        if solution == solution_old and possible_value == possible_value_old:
            solved = (min([sum(x) for x in zip(*solution)]) == max([sum(x) for x in zip(*solution)]) #Every column's sum are same
                      and min([sum(x) for x in solution]) == max([sum(x) for x in solution])         #Every row's sum are same
                      and min([sum(x) for x in zip(*solution)]) == sum(list(range(SIZE_N * SIZE_N + 1))) #The sum is 45 for column
                      and min([sum(x) for x in solution]) == sum(list(range(SIZE_N * SIZE_N + 1))))      #and for the row

            if solved:
                return solution
            else:
                result = make_guess(possible_value, solution)
                #If the guess does not produce anything then no solution
                #in the case it was a first guess, for stacked guess it means the 
                #preceding guess was wrong
                return result
```

```python
#Helper function to draw the puzzle
def print_sudoku(solution):
  for i, row in enumerate(solution):
    if i != 0 and i%SIZE_N == 0:
      print('\n--------------------------')
    else:
      print('\n')
    for j, col in enumerate(row):
      if j != 0 and j%SIZE_N == 0:
        print(' | ', end=' ')
      if col == 0:
        print('-', end=' ')
      else:
        print(col, end=' ')
  
  print('\n--------------------------\n')
```
